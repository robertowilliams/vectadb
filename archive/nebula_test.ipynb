# Nebula Graph Test Notebook
This notebook loads configuration, connects to a running Nebula Graph instance, and performs simple tests.

```python
import os
from nebula3.gclient.net import ConnectionPool
from nebula3.gclient.net import Session
from nebula3.Config import Config

NEBULA_HOST = os.getenv('NEBULA_HOST', '127.0.0.1')
NEBULA_PORT = int(os.getenv('NEBULA_PORT', '9669'))
NEBULA_USER = os.getenv('NEBULA_USER', 'root')
NEBULA_PASSWORD = os.getenv('NEBULA_PASSWORD', '')

print('Nebula host:', NEBULA_HOST, NEBULA_PORT)
```

```python
# Connect to Nebula
config = Config()
config.max_connection_pool_size = 10

pool = ConnectionPool()
ok = pool.init([(NEBULA_HOST, NEBULA_PORT)], config)
print('Pool init:', ok)

session = pool.get_session(NEBULA_USER, NEBULA_PASSWORD)
print('Session acquired')
```

```python
# Check spaces
result = session.execute('SHOW SPACES;')
for row in result.rows():
    print(row)
```

```python
# Test query (replace with your space)
try:
    session.execute('USE agent_registry;')
    r = session.execute('SHOW TAGS;')
    print('Tags:')
    for row in r.rows():
        print(row)
except Exception as e:
    print('Error:', e)
```
